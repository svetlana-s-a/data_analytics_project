#### ℹ️ **Практические рекомендации по выполнению проектов**

Работа аналитика в ноутбуке носит исследовательский и интерактивный характер: мы последовательно изучаем данные, делаем промежуточные выводы и принимаем решения о следующих шагах анализа. Python в данном процессе является инструментом, а ключевую роль играет аналитическая логика и интерпретация результатов.

Важно учитывать, что код для исследовательского анализа отличается от продакшн-кода. В ноутбуках не требуется писать универсальные и чрезмерно усложнённые скрипты, предназначенные для серверного исполнения. Главная задача - показать понятную логику исследования и обоснованные выводы.

**Раздел 1. Оформление аналитических проектов**

1. Читаемость кода

- Код должен быть простым и понятным, избегай избыточных конструкций и ненужного усложнения. 
- Логически разделяй код на небольшие блоки. 
- Предпочтение отдаётся кратким и прозрачным решениям, которые легко читать другим специалистам.

2. Структура ноутбука

- Ячейки типа Code используются только для кода и комментариев к логике кода.
- Названия этапов, пояснения, выводы и текстовые описания размещаются только в ячейках Markdown. 
- Каждый логический шаг анализа должен иметь заголовок и краткое пояснение. 
- Ноутбук должен отражать интерактивный процесс анализа: загрузка данных, исследование, выводы и переход к следующему шагу.

3. Импорты

- Все импорты располагаются в начале проекта. 
- Импорты не должны дублироваться в разных частях ноутбука без необходимости.

4. Простота решений

- Решения должны быть лаконичными: если задачу можно решить несколькими строками, не нужно писать сложные многоступенчатые конструкции.
- Следует избегать избыточных проверок, универсальных скриптов и чрезмерно длинных кодовых ячеек, которые ухудшают читаемость и не соответствуют интерактивной логике аналитической работы.

5. Использование материалов курса

- Решения должны опираться на инструменты и подходы, изученные в рамках спринта.
- Цель проекта - продемонстрировать освоение методов курса, а не использование сторонних готовых решений.
- При необходимости ревьюер может запросить дополнительные пояснения или выполнение отдельных шагов анализа для подтверждения понимания материала.

**Раздел 2. Общие требования к проектам**

1. Сохранение работы

- Регулярно сохраняй проект, так как несохранённый код может быть потерян при обновлении страницы.

2. Работоспособность проекта

- Перед отправкой проекта необходимо выполнить `Kernel → Restart & Run All`и убедиться, что все ячейки выполняются без ошибок.
- Если проект работает локально, но не работает на платформе Практикума, доработку необходимо выполнять на платформе. 
- В реальной работе также важно, чтобы проект корректно запускался на компьютере заказчика.

3. Работа с доработками

- Если проект отправлен на доработку, изменения вносятся в тот же файл.
- Комментарии ревьюера нельзя удалять и перемещать, иначе ревьюер вернёт последнюю версию проекта с восстановленными комментариями.
- Ответы на комментарии размещаются в Markdown-ячейках сразу после комментария.

4. Структура проекта

- Структура проекта должна соответствовать заданию и предоставленному шаблону выполнения проекта.
- Каждый этап анализа должен быть выделен отдельным блоком с пояснениями.
- Оформление и читаемость проекта являются обязательной частью работы аналитика и учитываются при проверке.

    - Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.

````python
# C помощью метода str.contains() найдем все значения "круглосуточно" в поле hours и выведем значения списком с помощью цикла
round_the_clock = df[df['hours'].str.contains('круглосуточно')]['hours'].unique()
for i, val in enumerate(round_the_clock, 1):
    print(f'{i}. {val}')
    
dayli = df[df['hours'].str.contains('24')]['hours'].unique()
for i, val in enumerate(dayli, 1):
    print(f"{i}. {val}")
````

Значение для указания работы 24/7 одно - "ежедневно, круглосуточно".
Создадим новый столбец `is_24_7` с обозначением того, что заведение работает 24/7

````python
# Создаем новый столбец и присваиваем значения с помощью функции:
def mode(value):
    if value == 'ежедневно, круглосуточно':
        return True
    else:
        return False

df['is_24_7'] = df['hours'].apply(mode)
df.info()
````
