### Данные

Для анализа используются предоставленные датасеты составленные на основе данных сервисов Яндекс Карты и Яндекс Бизнес за лето 2022 года. Информация, размещённая в сервисе Яндекс Бизнес, могла быть добавлена пользователями или найдена в общедоступных источниках - носит исключительно справочный характер.

### Описание rest_info.csv (содержит информацию о заведениях):
* `name` — название заведения
* `address` — адрес заведения
* `district` — административный район, в котором находится заведение, например Центральный административный округ
* `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»
* `hours` — информация о днях и часах работы
* `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
* `chain` — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
    - `0` — заведение не является сетевым
    - `1` — заведение является сетевым
* `seats` — количество посадочных мест

### Описание rest_price.csv (содержит информацию о среднем чеке в заведениях):
* `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
* `avg_bill` — хранит среднюю стоимость заказа в виде диапазона, например:
    - «Средний счёт: 1000–1500 ₽»
    - «Цена чашки капучино: 130–220 ₽»
    - «Цена бокала пива: 400–600 ₽»
* `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»:
    - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
    - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
    - Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
* `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»:
    - Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
    - Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
    - Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.

### Поиск явных и неявных дубликатов

Перед объединением датафреймов мы уже проверили на явыне дубликаты по полю `id` - явных дублей нет.

Выведем уникальные значения категориальных полей `category` и `district` для визуальной оценки:

```python
print('Уникальные значения столбца category:')
print(df['category'].unique(), '\n')
print('Уникальные значения столбца district:')
print(df['district'].unique(), '\n')
```

Поля `category` (8 уникальных значений) и `district` (9 уникальных значений) не содержат неявных дубликатов

Приводим строковые поля `name, address` к единому стилю написания, для дальнейшего поиска неявных дубликатов:

```python
for column in ['name', 'address', 'hours']:
    df[column] = df[column].str.lower()
    df[column] = df[column].str.strip()
```

Поиск дубликатов по паре name + address:

- Создать временное поле name + address
- Найти все строки с повторяющимся ключом
- Сортировать по `chain` по убыванию, чтобы строка с заведением "more poke" со значением 1 была первая и удалить дубль со значением 0

```python
df['name_add'] = df['name'] + '_' + df['address']

dup_mask = df['name_add'].duplicated(keep=False)

duplicates = df[dup_mask].sort_values(by=['name_add', 'chain'], ascending=[True, False]) 

print('Количество строк с повторяющимся значениями :', len(duplicates))

display(duplicates.head(10))
```

Удалим дубли и временный столбец:

```python
df = df.drop_duplicates(subset=['name_add'], keep='first')
df = df.drop(columns=['name_add'])
```

```python
df.info()
```

### Создаем столбец is_24_7 (заведение работает ежедневно и круглосуточно)

Необходимо найти значения в поле `hours`, по которому мы далее присовим True в новом столбце is_24_7

```python
round_the_clock = df[df['hours'].str.contains('круглосуточно')]['hours'].unique()
for i, val in enumerate(round_the_clock, 1):
    print(f'{i}. {val}')
    
dayli = df[df['hours'].str.contains('24')]['hours'].unique()
for i, val in enumerate(dayli, 1):
    print(f"{i}. {val}")
```

Значение для указания работы 24/7 одно - "ежедневно, круглосуточно".
Создадим новый столбец `is_24_7` с обозначением того, что заведение работает 24/7

```python
def mode(value):
    if value == 'ежедневно, круглосуточно':
        return True
    else:
        return False

df['is_24_7'] = df['hours'].apply(mode)
```

```python
df.info()
display(df.sample(5))
print('Значения в поле is_24_7:', df['is_24_7'].unique())
```
